```python
# prompt: dir of markdown to token calculator . bpe 

!pip install transformers sentencepiece

import os
from transformers import AutoTokenizer

def calculate_bpe_tokens(markdown_dir):
    """Calculates the number of BPE tokens in all markdown files within a directory.

    Args:
        markdown_dir: The path to the directory containing markdown files.

    Returns:
        A dictionary where keys are filenames and values are the number of BPE tokens.
        Returns an empty dictionary if no markdown files are found or the directory does not exist.
    """

    tokenizer = AutoTokenizer.from_pretrained("gpt2") # Example tokenizer; replace as needed
    token_counts = {}
    if not os.path.isdir(markdown_dir):
        print(f"Error: Directory '{markdown_dir}' not found.")
        return token_counts
    
    for filename in os.listdir(markdown_dir):
        if filename.endswith(".md"):
            filepath = os.path.join(markdown_dir, filename)
            try:
                with open(filepath, 'r', encoding='utf-8') as f:
                    markdown_content = f.read()
                    tokens = tokenizer.encode(markdown_content)
                    token_counts[filename] = len(tokens)
            except Exception as e:
                print(f"Error processing '{filename}': {e}")

    return token_counts

# Example usage (assuming your markdown files are in /md directory)
markdown_directory = "./files"
bpe_counts = calculate_bpe_tokens(markdown_directory)

if bpe_counts:
    for filename, count in bpe_counts.items():
        print(f"File: {filename}, BPE Tokens: {count}")

    total_tokens = sum(bpe_counts.values())
    print(f"\nTotal BPE Tokens across all files: {total_tokens}")
else:
    print("No token counts available.")
```
